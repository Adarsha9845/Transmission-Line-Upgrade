```python
import pandas as pd
```

```python
Path = "/content/drive/MyDrive/Dataset for EEE577/BranchData.xlsx"
```

```python
df = pd.read_excel(Path)
```

```python
df.head()
```

```python
bus_mapping = {**{i: i - 100 for i in range(101, 125)},
               **{i: i - 176 for i in range(201, 225)},
               **{i: i - 252 for i in range(301, 326)}}

# mapping technique
df['From Bus Updated'] = df['From Bus'].map(bus_mapping)
df['To Bus Updated'] = df['To Bus'].map(bus_mapping)

print(df)
```

```python
df.to_excel(Path, index=False)
```
